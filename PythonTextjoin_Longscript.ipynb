# READ ME

This script is written to textjoin(concatenate) alphanumeric texts into a long string, specifically for the master airway bill procedure for CNCB.</br>

#### Pre-requisites:</br> 
1. File is a CSV file
2. Python code is in the same folder as the CSV File

Therefore from the excel file that is generated from "Late_Tracker_Code.py", you need to export the 'order sn' sheet as an individual csv file. Then reference it in this notebook.
#### Use this when:
Textjoin does not work due to too many alphanumeric text 

```python
import pandas as pd
import subprocess
```

```python
def textjoin(file):
    df = pd.read_csv(file)
    lst = []
    for sn in df['order_sn']:
        "'" + sn + "'"
        lst.append(sn)
    out = ",".join(f"'{x}'" for x in lst)
    return out

    
```
